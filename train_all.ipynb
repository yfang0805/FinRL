```python
import pandas as pd
import os

from env import StockLearningEnv
import config

from stable_baselines3 import DDPG
from stable_baselines3 import A2C
from stable_baselines3 import PPO
from stable_baselines3 import TD3
from stable_baselines3 import SAC

import warnings

warnings.filterwarnings("ignore")
```

```python
train_df = pd.read_csv('./data_file/train.csv')

TECHNICAL_INDICATORS_LIST = [
    "boll_ub", "boll_lb", "rsi_20", "close_20_sma", "close_60_sma", "close_120_sma", \
    "macd", "volume_20_sma", "volume_60_sma", "volume_120_sma"
]

information_cols = TECHNICAL_INDICATORS_LIST + ["close", "day", "amount", "change", "daily_variance"]

ENV_TRAIN_PARAMS = {
    "initial_amount": 1e6,
    "hmax": 5000,
    "currency": '￥',
    "buy_cost_pct": 3e-3,
    "sell_cost_pct": 3e-3,
    "cache_indicator_data": True,
    "daily_information_cols": information_cols,
    "print_verbosity": 500,
    "patient": True,
}

e_train_gym = StockLearningEnv(df=train_df, **ENV_TRAIN_PARAMS)
```

```python
episode = 100000

model_A2C = A2C(policy='MlpPolicy', env=e_train_gym, **config.A2C_PARAMS)
model_A2C.learn(total_timesteps=episode)
model_A2C.save(os.path.join('train_file', "{}.model".format('A2C' + str(episode))))

model_DDPG = DDPG(policy='MlpPolicy', env=e_train_gym, **config.DDPG_PARAMS)
model_DDPG.learn(total_timesteps=episode)
model_DDPG.save(os.path.join('train_file', "{}.model".format('DDPG' + str(episode))))

model_PPO = PPO(policy='MlpPolicy', env=e_train_gym, **config.PPO_PARAMS)
model_PPO.learn(total_timesteps=episode)
model_PPO.save(os.path.join('train_file', "{}.model".format('PPO' + str(episode))))

model_SAC = SAC(policy='MlpPolicy', env=e_train_gym, **config.SAC_PARAMS)
model_SAC.learn(total_timesteps=episode)
model_SAC.save(os.path.join('train_file', "{}.model".format('SAC' + str(episode))))

model_TD3 = TD3(policy='MlpPolicy', env=e_train_gym, **config.TD3_PARAMS)
model_TD3.learn(total_timesteps=episode)
model_TD3.save(os.path.join('train_file', "{}.model".format('TD3' + str(episode))))
```
